# Gráficos FTIR

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
arquivos = os.listdir('FTIR')
```

## Gráficos juntos

```python
for i in range(len(arquivos)):
    df = pd.read_csv(f'FTIR/{arquivos[i]}',header=[3],sep='\t')
    df = df.reset_index()
    df = df.rename(columns={'index':'1/CM','##YUNITS=%T': '%T'})
    
    plt.plot(np.array(df['1/CM']),np.array(df['%T']))
    plt.xlabel('1/CM')
    plt.ylabel('%T')
```

## Gráficos separados

```python
for i in range(len(arquivos)):
    df = pd.read_csv(f'FTIR/{arquivos[i]}',header=[3],sep='\t')
    df = df.reset_index()
    df = df.rename(columns={'index':'1/CM','##YUNITS=%T': '%T'})
    
    plt.plot(np.array(df['1/CM']),np.array(df['%T']))
    plt.xlabel('1/CM')
    plt.ylabel('%T')
    plt.title(f'{arquivos[i]}')
    plt.show()
```
